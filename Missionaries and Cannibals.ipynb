```python
from IPython.display import clear_output
print("____Welcome to Missionaries and Cannibals Game____")

#Defining Functions
def printCurrentState(ml ,cl, mr, cr, boat_side):
    miss = '\U0001f482'
    cann = '\U0001f479'
    wave = '\U0001f30a'
    boat = '\U0001f6A2'
    
    print("The current state of the Game is: ")
    print(str(miss * ml) +  str(cann * cl), end = " " )
    if(boat_side == 'Right'):
        #print("|----B|", end = " " )
        print("|" + wave * 4 + boat + "|", end = " " )
    elif(boat_side == 'Left'):
        #print("|B----|", end = " " )
        print("|" + boat + wave * 4 + "|", end = " " )
    print(str(miss * mr) +  str(cann * cr))
    
def printInvalidMove():
    print("\n Invalid Move")
    
def boatRightToLeft(ml ,cl, mr, cr, missionaries, cannibals, boat_side):
    mr = mr - missionaries
    cr = cr - cannibals
    ml = ml + missionaries
    cl = cl + cannibals
    boat_side = 'Left'
    return ml ,cl, mr, cr, boat_side
    
def boatLeftToRight(ml ,cl, mr, cr, missionaries, cannibals, boat_side):
    mr = mr + missionaries
    cr = cr + cannibals
    ml = ml - missionaries
    cl = cl - cannibals
    boat_side = 'Right'
    return ml ,cl, mr, cr, boat_side
    
    
#Declaring variables
boat_side = 'Right'
mr = 3
cr = 3
ml = 0 
cl = 0
won = True

printCurrentState(ml ,cl, mr, cr, boat_side)
    
print("\nLet's the game begin!!!")

while True:
    missionaries = int(input("\n Enter the number of Missionaries: "))
    cannibals = int(input("\n Enter the number of Cannibals: "))
    
    if((ml + mr) <= 3 and (cl + cr) <= 3):
        if(not((missionaries == 1 and cannibals == 1) or (missionaries == 2 and cannibals == 0) 
           or (missionaries == 0 and cannibals == 2)  or (missionaries == 1 and cannibals == 0)
               or (missionaries == 0 and cannibals == 1))):
            printInvalidMove()
            continue

        #Checking the condition for Boat to be on the right or left
        if(boat_side == 'Right'):
            if(mr < missionaries or cr < cannibals):
                printInvalidMove()
                continue
            else:
                clear_output()
                ml ,cl, mr, cr, boat_side = boatRightToLeft(ml ,cl, mr, cr, missionaries, cannibals, "Left")
                if((ml + mr) <= 3 and (cl + cr) <= 3 and (ml >= 0) and (mr >= 0) and (cl >= 0) and (cr >= 0)):
                    printCurrentState(ml ,cl, mr, cr, boat_side)
                else:  
                    printInvalidMove()
        elif(boat_side == 'Left'):
            if((mr < missionaries or cr < cannibals) and (missionaries <= ml or cannibals <= cl)):
                printInvalidMove()
                continue
            else:
                clear_output()
                ml ,cl, mr, cr, boat_side = boatLeftToRight(ml ,cl, mr, cr, missionaries, cannibals, 'Right')
                if((ml + mr) <= 3 and (cl + cr) <= 3 and (ml >= 0) and (mr >= 0) and (cl >= 0) and (cr >= 0)):
                    printCurrentState(ml ,cl, mr, cr, boat_side)
                else:  
                    printInvalidMove()
                        
        if((mr < cr) and (mr > 0)):
            print("\nYou LOST THE GAME")
            break
        elif((ml < cl) and (ml > 0)):
            print("\nYou LOST THE GAME")
            break
        elif(ml == 3 and cl == 3):
            print("\nYOU HAVE WON THE GAME")
            break
    else:
        printInvalidMove()
        continue
        

print("\n The Game is Over. BYE BYE!")    
```
